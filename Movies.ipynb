```python
import pandas as pd
import json
from ast import literal_eval
```

```python
movies = pd.read_csv("/content/drive/MyDrive/movies_metadata.csv")
credits = pd.read_csv("/content/drive/MyDrive/credits.csv")
keywords = pd.read_csv("/content/drive/MyDrive/keywords.csv")
```

```python
movies.info()
```

```python
df_movies = movies.copy()
```

```python
df_movies['genres'] = df_movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df_movies['release_date'] = pd.to_datetime(df_movies['release_date'], errors='coerce').apply(lambda x: x.date() if pd.notnull(x) else x)

df_movies['production_countries'] = df_movies['production_countries'].fillna('[]').apply(literal_eval)
df_movies['production_countries'] = df_movies['production_countries'].apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

df_movies['production_companies'] = df_movies['production_companies'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

df_movies['spoken_languages'] = df_movies['spoken_languages'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
# Find non-integer values in the 'id' column
non_integer_ids = df_movies[df_movies['id'].apply(lambda x: not str(x).isdigit())]

# Display the problematic entries
non_integer_ids
```

```python
df_movies = df_movies[df_movies['id'].apply(lambda x: str(x).isdigit())]
```

```python
keywords['id'] = keywords['id'].astype('int')
credits['id'] = credits['id'].astype('int')
df_movies['id'] = df_movies['id'].astype('int')
```

```python
df_merge = df_movies.merge(credits, on='id')
df_merge = df_merge.merge(keywords, on='id')
```

```python
df_merge
```

```python
df_merge['cast'] = df_merge['cast'].apply(literal_eval)
df_merge['crew'] = df_merge['crew'].apply(literal_eval)

df_merge
```
