```python
df_movies = movies.copy()
```

The ID is an integer value that serves as a unique identifier for each entry in the dataset. We need to ensure that the ID used for merging the dataframes is correct.

```python
df_movies[df_movies['id'].apply(lambda x: not str(x).isdigit())]
```

Remove the ID that contains incorrect data.

```python
df_movies = df_movies[df_movies['id'].apply(lambda x: str(x).isdigit())]
```

Transform the ID type for each dataframe.

```python
keywords['id'] = keywords['id'].astype('int')
credits['id'] = credits['id'].astype('int')
df_movies['id'] = df_movies['id'].astype('int')
```

Merge dataframe

```python
df_merge = df_movies.merge(credits, on='id')
df_merge = df_merge.merge(keywords, on='id')
```

```python
df_merge.head()
```

We will extract the column containing the data from the JSON and convert it into a list. Select the name from this column.

```python
df_merge['genres'] = df_merge['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

df_merge['production_countries'] = df_merge['production_countries'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

df_merge['production_companies'] = df_merge['production_companies'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])

df_merge['spoken_languages'] = df_merge['spoken_languages'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

Select the actor from the cast.

```python
df_merge['cast'] = df_merge['cast'].apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

For crew, we select the director.

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df_merge['crew'] = df_merge['crew'].apply(literal_eval).apply(get_director)
```

For keywords, take out the phrases from the json type.

```python
df_merge['keywords'] = df_merge['keywords'].apply(literal_eval).apply(lambda x: [item['name'] for item in x])
```

Take out the collection from 'belongs_to_collection'.

```python
df_merge['belongs_to_collection'] = df_merge['belongs_to_collection'].fillna('[]').apply(lambda x: literal_eval(x)['name'] if isinstance(literal_eval(x), dict) else np.nan)
```
